```python
#lets predict the central pressure('CP') with the help of the MSWS, tlr30, tsr30, Loc and Lac
#loading the model while we are using the Decision tree classifier 
#we are alos load and do feature selction for the model fitting
#X_t = ['Lac', 'Loc', 'MSWS', 'tlr', 'tsr', 'tlr30', 'tsr30']
#X = sample[['Lac', 'Loc', 'MSWS', 'tlr', 'tsr', 'tlr30', 'tsr30']]
X = sample.iloc[:, [2, 3, 5]]
Y = sample['cp']
X
```

```python
#model loading and deployment
tree = DecisionTreeClassifier()
tree.fit(X, Y)
```

```python
prediction = tree.predict([[73, 1560, 45]])
print("The Central Pressure is : ", prediction)
```
